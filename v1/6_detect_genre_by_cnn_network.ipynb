```python
import json
import librosa
import numpy as np
import tensorflow as tf
import tensorflow.keras as krs
from tensorflow.keras.models import load_model
```

```python
model = load_model("my_model_v2.h5")
```

```python
with open("data.json", "r") as fp:
    data = json.load(fp)

n = 20
m = 100
data_part = {

        "mapping": data["mapping"][: m],
        "mfcc": data["mfcc"][: m],
        "labels": data["labels"][: m], 
}
data_part
```

```python
WAV_PATH = "9_genres_record.wav"

SAMPLE_RATE  = 8000
WAV_DURATION = 45 
WAV_PIECE_LEN = 0.3
NUM_SEGMENTS = int(WAV_DURATION / WAV_PIECE_LEN)


# mfcc
N_FFT = 256
N_MFCC = 13
HOP_LENGTH = 64

```

```python
# take next segment of audio file
# make mfcc 
# evaluate model
# print status 

results = []

signal, sr = librosa.load(WAV_PATH, sr=SAMPLE_RATE)

signal_duration = int(len(signal) / sr)
pieces = int(signal_duration / WAV_PIECE_LEN)

for i in range(0, pieces):
    offset = i * WAV_PIECE_LEN
    sub_signal, sr = librosa.load(WAV_PATH, offset=offset, duration=WAV_PIECE_LEN, sr=SAMPLE_RATE)

    mfcc = librosa.feature.mfcc(
        y=sub_signal, 
        sr=SAMPLE_RATE,
        n_fft=N_FFT,
        n_mfcc=N_MFCC, 
        hop_length=HOP_LENGTH).T
    mfcc_as_list = np.array(mfcc.tolist())


    x = mfcc_as_list[ np.newaxis, ...]
    prediction = model.predict(x)
    p_index = np.argmax(prediction, axis=1)[0]
    mapping = data_part["mapping"][p_index]
    results.append(mapping)
    
    

    




```
