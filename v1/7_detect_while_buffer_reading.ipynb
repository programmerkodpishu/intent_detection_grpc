```python
import time
import librosa
import numpy as np
import sounddevice as sd
from IPython.display import Audio
from tensorflow.keras.models import load_model

```

```python
SAMPLE_RATE = 8000
PIECE_LEN = 0.3 
SAMPLE_RATES_IN_PIECE = SAMPLE_RATE * PIECE_LEN

WAV_PATH = "9_genres_record.wav"
WAV_LEN = 45


# mfcc
N_FFT = 256
N_MFCC = 13
HOP_LENGTH = 64
```

```python
Audio(WAV_PATH)
```

```python
GENRES = [
        'disco',
        'blues',
        'rock',
        'pop',
        'metal',
        'country',
        'hiphop',
        'classical',
        'reggae'
        ]
model = load_model("my_model_v2.h5")
```

```python
def get_next_signal():
    for i in range(0, int(WAV_LEN/PIECE_LEN)):
        offset = int(i * PIECE_LEN)
        signal, sr = librosa.load(WAV_PATH, mono=True, offset=offset, duration=PIECE_LEN, sr=SAMPLE_RATE)
        yield signal

def build_mfcc(signal):
    mfcc = librosa.feature.mfcc(y=signal,sr=SAMPLE_RATE,n_fft=N_FFT,n_mfcc=N_MFCC,hop_length=HOP_LENGTH).T
    return np.array(mfcc.tolist())[np.newaxis, ...]


def get_recognized_genre(model, mfcc):
    prediciton = model.predict(mfcc)
    max_arg_index = np.argmax(prediciton, axis=1)[0]
    return GENRES[max_arg_index]


def play_signal_and_print_genre(signal, genre):
    sd.play(signal, samplerate=SAMPLE_RATE)
    print(genre)
    time.sleep(1)
```

```python
it = get_next_signal()


while True:
    try:
        signal = next(it)
        mfcc = build_mfcc(signal)
        genre = get_recognized_genre(model, mfcc)
        play_signal_and_print_genre(signal, genre)
    except StopIteration:
        break
    

```
