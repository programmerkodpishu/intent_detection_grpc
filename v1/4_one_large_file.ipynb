```python
import os
import librosa
import soundfile as sf
from IPython.display import Audio
```

```python
PATH = "../Data/genres_one_sample_each/"
SAMPLE_RATE = 22050
PART_LEN_SEC = 5
PART_LEN_SAMPLES = int(PART_LEN_SEC * SAMPLE_RATE)
print(PART_LEN_SAMPLES)
lst = os.listdir(PATH)
main_sgnal = []
for f in lst:
    signal, sr = librosa.load(os.path.join(PATH, f), sr=SAMPLE_RATE)
    main_sgnal.extend(signal[:PART_LEN_SAMPLES])
```

```python
Audio(main_sgnal, rate=SAMPLE_RATE)
```

```python
sf.write("9_genres_record.wav", data=main_sgnal, samplerate=SAMPLE_RATE)
```
